# 03 - Interference Basics

This notebook focuses on one key idea: **phase can be hidden at first and become visible after more gates**.

Prediction-first workflow: before each run, write what you expect.

```python
from qiskit import QuantumCircuit, transpile
from qiskit.providers.basic_provider import BasicSimulator
```

```python
def run_counts(circuit: QuantumCircuit, shots: int = 1024) -> dict[str, int]:
    backend = BasicSimulator()
    compiled = transpile(circuit, backend)
    return backend.run(compiled, shots=shots).result().get_counts(compiled)

def probs(counts: dict[str, int]) -> dict[str, float]:
    total = sum(counts.values())
    return {k: v / total for k, v in sorted(counts.items())}
```

## Circuit A: H then measure

Prediction: close to 50/50 between `0` and `1`.

```python
# QuantumCircuit(num_qubits, num_classical_bits)
a = QuantumCircuit(1, 1)

# H on qubit 0 creates equal superposition.
a.h(0)

# Measure qubit 0 into classical bit 0.
a.measure(0, 0)

a.draw(output='text')
```

```python
a_counts = run_counts(a, shots=1024)
a_counts, probs(a_counts)
```

## Circuit B: H then Z then measure

Prediction: still close to 50/50 in direct Z-basis measurement.

```python
b = QuantumCircuit(1, 1)
b.h(0)

# Z changes phase, not direct bit probability in this measurement basis.
b.z(0)
b.measure(0, 0)
b.draw(output='text')
```

```python
b_counts = run_counts(b, shots=1024)
b_counts, probs(b_counts)
```

## Circuit C: H then H then measure

Prediction: mostly `0`, because H-H acts like identity.

```python
c = QuantumCircuit(1, 1)
c.h(0)
c.h(0)
c.measure(0, 0)
c.draw(output='text')
```

```python
c_counts = run_counts(c, shots=1024)
c_counts, probs(c_counts)
```

## Circuit D: H then Z then H then measure

Prediction: mostly `1`, because H-Z-H behaves like X on `|0>`.

```python
d = QuantumCircuit(1, 1)

# First H creates superposition.
d.h(0)

# Z injects a relative phase sign change.
d.z(0)

# Final H converts phase information into measurable population.
d.h(0)

d.measure(0, 0)
d.draw(output='text')
```

```python
d_counts = run_counts(d, shots=1024)
d_counts, probs(d_counts)
```

## Shot Stability

Compare 100 shots vs 10,000 shots for Circuit A.

```python
a_100 = run_counts(a, shots=100)
a_10000 = run_counts(a, shots=10000)

print('100 shots:', a_100, probs(a_100))
print('10000 shots:', a_10000, probs(a_10000))
```

- Circuit D shows how a phase difference can be converted into a bit-value difference.
